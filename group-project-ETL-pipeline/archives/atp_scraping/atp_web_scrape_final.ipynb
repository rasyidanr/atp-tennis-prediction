### Scrap Player's Ranks
Get the ranking of top 2k players

Page: https://www.atptour.com/en/rankings/singles?rankRange=1-5000&rankDate=2022-03-07

```python
# Prepare file 

import csv

header = ['player', 'rank', 'age', 'point', 'tournament_played', 'point_drop','next_best']

with open('tennis_ranking_data.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)
```

```python
# Open browser
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get('https://www.atptour.com/en/rankings/singles?rankRange=1-5000&rankDate=2022-03-07')
```

```python
# Run selenium to pull data

with open('tennis_ranking_data.csv', 'a', encoding='UTF8', newline='') as f:
    
    writer = csv.writer(f)
    
    for i in tqdm(range(1,2001)):

        player = WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="player-rank-detail-ajax"]/tbody/tr['+str(i)+']/td[4]/span/a')))
        rank=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="player-rank-detail-ajax"]/tbody/tr['+str(i)+']/td[1]')))
        age=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="player-rank-detail-ajax"]/tbody/tr['+str(i)+']/td[5]')))
        point=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="player-rank-detail-ajax"]/tbody/tr['+str(i)+']/td[6]/a')))
        tourn_played=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="player-rank-detail-ajax"]/tbody/tr['+str(i)+']/td[7]/a')))
        point_drop=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="player-rank-detail-ajax"]/tbody/tr['+str(i)+']/td[8]')))
        next_best=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="player-rank-detail-ajax"]/tbody/tr['+str(i)+']/td[9]')))
        
        data=[player.text,rank.text,age.text,point.text,tourn_played.text,point_drop.text,next_best.text]
        writer.writerow(data)
```

### Scrap Leaderboard

Page: https://www.atptour.com/en/stats/leaderboard?boardType=serve&timeFrame=Career&surface=all&versusRank=all&formerNo1=false

```python
# Prepare file 

import csv

header = ['player', 'serve_rating', '1st_serve', '1st_serve_points_won', 
          '2nd_serve_points_won', 'service_games_won_percent','avg_aces_per_match','avg_double_faults_per_match']

with open('tennis_leaderboard_serve.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)
```

```python
# Open browser
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get('https://www.atptour.com/en/stats/leaderboard?boardType=serve&timeFrame=Career&surface=all&versusRank=all&formerNo1=false')
```

```python
# Loop through the list 

with open('tennis_leaderboard_serve.csv', 'a', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    for j in tqdm(range(1, 201)):
        player = WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="leaderboardTable"]/tr['+str(j)+']/td[2]/div/a')))
        serve_rating = WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="leaderboardTable"]/tr['+str(j)+']/td[3]')))            
        serve_1st=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="leaderboardTable"]/tr['+str(j)+']/td[4]')))  
        serve_won_1st=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="leaderboardTable"]/tr['+str(j)+']/td[5]'))) 
        serve_won_2nd=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="leaderboardTable"]/tr['+str(j)+']/td[6]'))) 
        service_games_won=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="leaderboardTable"]/tr['+str(j)+']/td[7]'))) 
        avg_aces_per_match=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="leaderboardTable"]/tr['+str(j)+']/td[8]'))) 
        avg_double_fault_per_match=WebDriverWait(browser, 20).until(expected_conditions.visibility_of_element_located((By.XPATH, '//*[@id="leaderboardTable"]/tr['+str(j)+']/td[9]')))
                
        data=[player.text,serve_rating.text,serve_1st.text,serve_won_1st.text,serve_won_2nd.text,
              service_games_won.text,avg_aces_per_match.text,avg_double_fault_per_match.text]
        writer.writerow(data)
```

### Scrap Tournament Page

Page: https://www.atptour.com/en/scores/results-archive?year=2022'

```python
import csv

header = ['tournament_name', 'tournament_location','tournament_date',
          'tournament_draw', 'tournament_surface', 'tournament_financial_commitment']

with open('tournament_details_scrape.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)
```

```python
# Open browser
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get('https://www.atptour.com/en/scores/results-archive?year=2022')
```

```python
all_offers = browser.find_elements_by_class_name("tourney-result")

data = []

for offer in all_offers:
    # uses `offer` instead `driver` to search only in this one offer
    # uses `element` instead of `elements` to search only one value

    try:
        title_content = offer.find_element_by_class_name("title-content").text.strip()
    except Exception as ex:
        #print('[Exception] company:', ex)
        title_content = 'NAN'

    tournament_name = offer.find_element_by_class_name("tourney-title").text.strip()
    tournament_location = offer.find_element_by_class_name("tourney-location").text.strip()
    tournament_date = offer.find_element_by_class_name("tourney-dates").text.strip()
    #tournament_draw = offer.find_element_by_class_name("item-details").text.strip()
    #tournament_surface = offer.find_element_by_class_name("tourney-details").text.strip()
    tournament_financial_commitment = offer.find_element_by_class_name("tourney-details.fin-commit").text.strip()
    
    #looping through the tourney_detail class to get each of tournament draw and surface
    #as these have same class names and we were unable to get values by using class names
    #as done above for other variables.
    all_offers1 = browser.find_elements_by_class_name("tourney-details")
    count=0
    for offer1 in all_offers1:
        count=count+1
        if count==1:
            tournament_draw = offer1.text
        elif count ==2:
            tournament_surface = offer1.text
        else:
            break
            
            
    data.append([tournament_name,tournament_location, tournament_date, tournament_draw, tournament_surface,tournament_financial_commitment])
    
with open('tournament_details_scrape.csv', 'a', encoding='UTF8', newline='') as f:
    #for row in data:
    write = csv.writer(f)
    write.writerows(data)    
```

```python
# Define class for match score
all_data = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[2]/div[1]/div/div[5]/div/table')
players = all_data.find_elements(by=By.CLASS_NAME, value='day-table-name')
scores = all_data.find_elements(by=By.CLASS_NAME, value='day-table-score')
```

```python
# Create dataframe for player's details
match_player = dict()
match_score = dict()

match_player_single = pd.DataFrame()
match_score_single = pd.DataFrame()
```

```python
# Open browser single
driver.get(web_single)
time.sleep(10)
```

```python
# Define class for tournament  
tournament_data = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[2]/div[1]/div/div[2]/div/div/div/div/table')
tournament_title = tournament_data.find_elements(by=By.CLASS_NAME, value='tourney-title')
tournament_location = tournament_data.find_elements(by=By.CLASS_NAME, value='tourney-location')
tournament_dates = tournament_data.find_elements(by=By.CLASS_NAME, value='tourney-dates')
tournament_prize = tournament_data.find_elements(by=By.CLASS_NAME, value='item-value')
```

```python
# Fetch tournament details

for title in tournament_title:
    title=title.text

for loc in tournament_location:
    location=loc.text
    
for date in tournament_dates:
    date=date.text

for prize in tournament_prize:
    prize=prize.text
```
